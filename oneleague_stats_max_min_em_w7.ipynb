```python
def f(x, y):
  return x**2 + y**2

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
```

```python
ax = plt.axes(projection = '3d')
# ax.plot3D(x, y, f(x,y))
ax.plot_surface(X, Y, f(X,Y))
```

```python
ax = plt.axes(projection = '3d')
# ax.plot3D(x, y, f(x, y))
ax.plot_surface(X, Y, f(X, Y), alpha = 0.3)
ax.contour3D(X, Y, f(X, Y), zdir = 'x', offset = -2 )
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 1000,  random_state = 22)
X.shape
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.mixture import GaussianMixture
```

```python
mix = GaussianMixture(n_components = 3)
```

```python
mix.fit(X)
```

```python
means = mix.means_
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
plt.plot(means[:,0], means[:,1], 'ro')
```

```python
mix.predict(X)
```
